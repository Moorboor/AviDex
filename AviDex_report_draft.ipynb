This notebook documents the rationale, process, findings and above all the code for our group project (Main van Do and Lay Vu). The project name *AviDex* is directly inspired by the fictional *Pokédex*, a register attributing an index to each known Pokémon species through which its physiology, habits, evolution, etc. can be looked up.  
(Un)fortunately, our project does not span thousands of species yet. On the one hand each of us could only spend so time on the task (~15h per person). On the other hand we had to deal with data constraints.

To be specific, we diverged from the suggested way of completing the project and instead to came up with our own dataset. This involved finding a topic of interest to both of us and developing a sensible research question/angle tied to that topic.
