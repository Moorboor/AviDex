# Introduction
This notebook documents the rationale, process, findings and above all the code for our final project as a group (Main van Do and Lay Vu) in the course "Scientific Python" (WS 2025/26, University of Osnabrück).  

Our project name *AviDex* is directly inspired by the fictional *Pokédex*, a register attributing an index to each known Pokémon species through which its "stats" (scores) for attack, health, defense, etc. can be looked up at a glance. (cf. [1]) Instead of Pokémon, the entry subjects in our index are songbirds and characteristics of their songs. In other words, we started to build an **avi**an in**dex**. Also unlike the Pokédex, our project does not span thousands of species (yet), in fact there are only two. On the one hand this had to do with how each of us could only spend so much time on the task, limited to around 15h per person. On the other hand we had to deal with data constraints.

## Finding a topic, scope and research aim
The latter came up in part because we diverged from the suggested way of doing the project and wanted to create our own dataset. The data constraints arose further due to our chosen topic of interest being birds.
[To get there, we had to find a topic of interest to both of us and then enough data on that topic to develop a sensible research question/angle tied to that topic.]
It turned out that Maik's grandfather, an ornithologist aka bird researcher, had recorded many different kinds of bird calls and songs over the span of half a century—and both of us find birds pretty awesome. We were granted access to the data and excited to make sense of bird vocalisations. However, not all of the audio recordings (almost 4,000 in total) were of interest to us, since we wanted to focus on bird songs in particular, limiting us to the represented subset of songbirds.  
Based on this exploration of the available data by ourselves, we initially planned to compare various birds in terms of the frequencies occurring in their songs and their verse durations, etc. This is also what we relayed to our tutor in the first project meeting.  

Afterwards, however, we consulted Maik's grandfather and came to agree with his suggestion that it might be more interesting to not simply compare random birds (of different families at that) but instead to analyse the songs of closely related species within the same family. That way we'd be able to find more meaningful commonalities and differences. Since ornithologists mostly work qualitatively and thus usually do not have numerous recordings of multiple individual birds of a species, we had to choose two or three related species for which Maik's grandfather would be able to supply a somewhat sufficient amount of recordings. Ultimately, we decided to compare to kinds of European robins: *Erithacus rubecula rubecula* (hereafter *rubecula*) and *Erithacus rubecula superbus* (hereafter *superbus*), with the latter being a distinct subspecies found on the central Canarian islands of Tenerife and Gran Canaria [2].
