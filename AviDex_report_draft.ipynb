Ultimately, we decided to compare to kinds of European robins: *Erithacus rubecula rubecula* (hereafter *rubecula*) and *Erithacus rubecula superbus* (hereafter *superbus*), with the latter being a distinct subspecies found on the central Canarian islands of Tenerife and Gran Canaria [2].

### 3.2 Mel spectograms
Sound is commonly visualized with spectograms. All spectograms are computed by analyzing patterns of sounds. They are often computed by Fourier transformations, while other transformations are less popular. When used in bioacoustics, Ornithologists usually refer to denoised spectograms as sonagrams - highlighting the bird's singing. Here, we used a special form of spectograms called mel spectograms Librosa *https://librosa.org/*. Mel spectograms use a special frequency scale that matches frequencies of human auditory experience.

### 3.3 Segmentation
For analyzing songs in songbirds, it is essential to differentiate between strophes and pauses. We explored a variety of algorithms, including Librosa's inhouse detect onsets [Figure 1], mean of decibles $\mu(t) = \langle f_{mel}(t)\rangle$ with computing a binary array $\mu(t) > \langle\mu(t)\rangle$ and maximum of mel frequencies coupled with a moving average $$\mathrm{MA}(t) = \mathrm{conv}(\max\left(f_{mel}(t)\right)$$ to mark robust segments of strophes and pauses [Figure 2]. Subsequently, we used a customized form of **erosion** and **dilation** to eliminate noisy false alarms. After we retrieved binary time series of strophes, we used the **first differentiation** to obtain the start and end of strophes [Figure 3].
